```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas.io.sql as psql
from scipy.cluster.vq import kmeans,vq, whiten
%matplotlib inline
```

```python
# http://www12.statcan.gc.ca/census-recensement/2011/dp-pd/prof/details/page.cfm?Lang=E&Geo1=FED2013&Code1=24028&Geo2=PR&Code2=24&Data=Count&SearchText=Hochelaga&SearchType=Begins&SearchPR=01&B1=All&GeoLevel=PR&GeoCode=028&TABID=1
# download Federal electoral districts (2013 representation order)
df = pd.read_csv("./data/98-316-XWE2011001-511.csv", low_memory=False, skiprows=1)
```

```python
df.head()
```

```python
df["Topic_Characteristic"] = df.Topic + ' ' + df.Characteristic
```

```python
census = df.pivot_table(index=['Geo_Code', 'Prov_Name', 'FED_Name'], columns='Topic_Characteristic', values='Total')
census.reset_index(inplace=True)
census.head()
```

```python
# Compute male ratio
ratio_male = df.ix[df.Characteristic=='Total population by age groups',['Geo_Code', 'Total', 'Male']]
ratio_male['ratio_male'] = ratio_male.Male/ratio_male.Total

census = pd.merge(census, ratio_male[['Geo_Code', 'ratio_male', 'Total']], on='Geo_Code', how='left')
```

```python
# Plotting the resulting variable
census['ratio_male'].hist()
plt.show()
plt.boxplot(census['ratio_male'])
plt.show()
```

```python
census['Total by age'] = (census["Age characteristics    0 to 4 years"] +
    census["Age characteristics    5 to 9 years"] +
    census["Age characteristics    10 to 14 years"] +
    census["Age characteristics    15 to 19 years"] +
    census["Age characteristics    20 to 24 years"] +
    census["Age characteristics    25 to 29 years"] +
    census["Age characteristics    30 to 34 years"] +
    census["Age characteristics    35 to 39 years"] +
    census["Age characteristics    40 to 44 years"] +
    census["Age characteristics    45 to 49 years"] +
    census["Age characteristics    50 to 54 years"] +
    census["Age characteristics    55 to 59 years"] +
    census["Age characteristics    60 to 64 years"] +
    census["Age characteristics    65 to 69 years"] +
    census["Age characteristics    70 to 74 years"] +
    census["Age characteristics    75 to 79 years"] +
    census["Age characteristics    80 to 84 years"])

census['Ratio Age 0 to 14'] = (census["Age characteristics    0 to 4 years"] +
                                census["Age characteristics    5 to 9 years"] +
                                census["Age characteristics    10 to 14 years"]) / census['Total by age']

census['Ratio Age 15 to 34'] = (census["Age characteristics    15 to 19 years"] +
                                census["Age characteristics    20 to 24 years"] +
                                census["Age characteristics    25 to 29 years"] + 
                                census["Age characteristics    30 to 34 years"]) / census['Total by age']

census['Ratio Age 35 to 54'] = (census["Age characteristics    35 to 39 years"] +
                                census["Age characteristics    40 to 44 years"] +
                                census["Age characteristics    45 to 49 years"] + 
                                census["Age characteristics    50 to 54 years"]) / census['Total by age']

census['Ratio Age 55 to 84'] = (census["Age characteristics    55 to 59 years"] +
                                census["Age characteristics    60 to 64 years"] +
                                census["Age characteristics    65 to 69 years"] + 
                                census["Age characteristics    70 to 74 years"] +
                                census["Age characteristics    75 to 79 years"] +
                                census["Age characteristics    80 to 84 years"]) / census['Total by age']

census['Ratio Age 85+'] = census['Age characteristics    85 years and over'] / census['Total by age']
```

```python
age_cols = ['Ratio Age 0 to 14', 'Ratio Age 15 to 34', 'Ratio Age 35 to 54', 'Ratio Age 55 to 84', 'Ratio Age 85+']
for col in age_cols: 
    census[col].hist()
    plt.show()
```

```python
marital_status_list = ['Marital status       Divorced',
   'Marital status       Living common law',
   'Marital status       Married (and not separated)',
   'Marital status       Separated',
   'Marital status       Single (never legally married)',
   'Marital status       Widowed']

census['Total by Marital status'] = 0

for col in marital_status_list:
    census['Total by Marital status'] += census[col]

for col in marital_status_list:
    census['Ratio ' + col] = census[col] / census['Total by Marital status']
```

```python
family_size_list = ['Family characteristics    Size of census family: 2 persons',
       'Family characteristics    Size of census family: 3 persons',
       'Family characteristics    Size of census family: 4 persons',
       'Family characteristics    Size of census family: 5 or more persons']

census['Total by Family Size'] = 0

for col in family_size_list:
    census['Total by Family Size'] += census[col]

for col in family_size_list:
    census['Ratio ' + col] = census[col] / census['Total by Family Size']
```

```python
first_language_list = ['First official language spoken \xa0 English',
       'First official language spoken \xa0 English and French',
       'First official language spoken \xa0 French',
       'First official language spoken \xa0 Neither English nor French']

census['Total by First official language'] = 0

for col in first_language_list:
    census['Total by First official language'] += census[col]

for col in first_language_list:
    census['Ratio ' + col.replace('\xa0', '')] = census[col] / census['Total by First official language']
```

```python
cols_kmeans = ['Ratio ' + s for s in marital_status_list] + \
                ['Ratio First official language spoken  English','Ratio First official language spoken  English and French'
                 ,'Ratio First official language spoken  French','Ratio First official language spoken  Neither English nor French'
                 , 'ratio_male']
census_kmeans = census[cols_kmeans]
```

```python
cols_kmeans
```

```python
p = pd.tools.plotting.scatter_matrix(census_kmeans, alpha=0.2, figsize=(16,16))
```

```python
k_range = range(1, 11) 

features = census_kmeans.values
whitened = whiten(features)

k_wcss = list()

for k in k_range:
    centroids, _ = kmeans(whitened, k)
    code, dist = vq(whitened, centroids)
    k_wcss.append(dist.sum())
```

```python
plt.ylim(0, 2000)
plt.plot(k_wcss)
```

```python
k=3

centroids, _ = kmeans(whitened, k)
code, dist = vq(whitened, centroids)
```

```python
census_kmeans['cluster'] = code
census_kmeans['cluster_distance'] = dist
```
