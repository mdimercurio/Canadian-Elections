```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas.io.sql as psql
from scipy.cluster.vq import kmeans,vq, whiten
%matplotlib inline
```

```python
df = pd.read_csv("./data/TRANSPOSITION_338FED.csv", low_memory=False, skiprows=4)
```

```python
df.head()
```

```python
'Conservative votes', 'NDP-New Democratic Party votes','Liberal votes', 'Bloc Qu\xe9b\xe9cois votes', 'Green Party votes','Other votes'
```

```python
votes = pd.DataFrame(df.groupby(['2013 FED Name'])['Conservative votes', 'NDP-New Democratic Party votes',
       'Liberal votes', 'Bloc Qu\xe9b\xe9cois votes', 'Green Party votes',
       'Other votes'].sum()).reset_index()

votes.head()
```

```python
votes_col = ['Conservative votes', 'NDP-New Democratic Party votes',
       'Liberal votes', 'Bloc Qu\xe9b\xe9cois votes', 'Green Party votes',
       'Other votes']

votes['Total'] = 0

for col in votes_col:
    votes['Total'] += votes[col]

for col in votes_col:
    votes['Ratio ' + col] = votes[col] / votes['Total']
```

```python
votes.head()
```

```python
cols_kmeans = ['Ratio ' + s for s in votes_col]
votes_kmeans = votes[cols_kmeans]
```

```python
votes_kmeans.rename(columns={'Ratio Bloc Qu\xe9b\xe9cois votes':'Ratio Bloc Quebecois votes'}, inplace=True)
p = pd.tools.plotting.scatter_matrix(votes_kmeans, alpha=0.2, figsize=(16,16))
```

```python
k_range = range(1, 11) 

features = votes_kmeans.values
whitened = whiten(features)

k_wcss = list()

for k in k_range:
    centroids, _ = kmeans(whitened, k)
    code, dist = vq(whitened, centroids)
    k_wcss.append(dist.sum())
```

```python
#plt.ylim(500, 1400)
plt.plot(k_wcss)
```

```python
k=3

centroids, _ = kmeans(whitened, k)
code, dist = vq(whitened, centroids)
votes_kmeans['cluster'] = code
votes_kmeans['cluster_distance'] = dist
votes_kmeans.head()
```

```python
# number of FED by cluster
votes_kmeans.groupby('cluster').count().ix[:,0]
```

```python
votes_kmeans['FED_Name'] = votes['2013 FED Name']
```
