```python
census = df.ix[df.Characteristic=='Population in 2011', ['Geo_Code', 'Prov_Name', 'FED_Name']]
```

```python
test = df[df.Characteristic=='Total population by age groups'][['Geo_Code', 'Total', 'Male', 'Female']]
test['male_ratio'] = test.Male / test.Total
test = test.drop(['Male', 'Female'], axis=1)
```

```python
census = pd.merge(census, test, on='Geo_Code')
```
